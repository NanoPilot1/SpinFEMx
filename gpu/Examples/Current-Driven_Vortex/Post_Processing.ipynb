```python
import numpy as np
import matplotlib.pyplot as plt

# We compare the results obtained using the hybrid FEM/BEM approach (Lindholm) and the fast multipole method.

data1 = np.loadtxt("STT_lind_gpu/log_stt_gpu_bdf.txt", skiprows=1)

plt.plot(data1[:,0],data1[:,2], color='black',linestyle='dotted', label="m_x")
plt.plot(data1[:,0],data1[:,3], color='black', linestyle='dotted', label="m_x")
plt.plot(data1[:,0],data1[:,4], color='black', linestyle='dotted', label="m_x")

data2 = np.loadtxt("STT_fmm_gpu/log_stt_gpu_bdf.txt", skiprows=1)
plt.plot(data2[:,0],data2[:,2], color='red', label="m_x fmm")
plt.plot(data2[:,0],data2[:,3], color='blue', label="m_y fmm")
plt.plot(data2[:,0],data2[:,4], color='green', label="m_z fmm")


plt.xlabel("t (ns)")
plt.ylabel("m")
plt.ylim([-0.4,0.3])
plt.legend()
plt.tight_layout()
plt.show()
```
